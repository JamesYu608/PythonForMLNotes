```python
# The normal imports
import numpy as np
from numpy.random import randn
import pandas as pd

# Import the stats library from numpy
from scipy import stats

# These are the plotting modules adn libraries we'll use:
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# Command so that plots appear in the iPython Notebook
%matplotlib inline
```

# Combining Plot Styles

```python
dataset = randn(100)
```

```python
# Default: Hist + KDE
sns.distplot(dataset, bins=25)
```

```python
# Rug + KDE
sns.distplot(dataset, bins=25, rug=True, hist=False)
```

```python
# 使用dict分別設定各個plot的attributes
sns.distplot(dataset, bins=25,
             kde_kws={'color': 'indianred', 'label': 'KDE PLOT'},
             hist_kws={'color': 'blue', 'label': 'HIST'})
```

### seaborn可以吃pandas的data

```python
from pandas import Series

ser1 = Series(dataset, name='My Data')
ser1
```

```python
sns.distplot(ser1, bins=25)
```

# Box and Violin Plots

```python
data1 = randn(100)
data2 = randn(100)
```

### Box Plot

```python
sns.boxplot(data=[data1, data2])
```

```python
# 不顯示outliers
# Box plot的outliers: 超過一定範圍(Q1 - 1.5* IQR...)的值，見box plot的說明
sns.boxplot(data=[data1, data2], whis=np.inf) # np.int表示無限大
```

```python
# orient: v / h
sns.boxplot(data1, orient='h')
```

# Violin Plot
Box plot只是展示了分位數的位置，violin plot可以展示任意位置的"密度"

```python
# Normal Dist
data1 = stats.norm(0, 5).rvs(100) # rvs(n): 產生n個點

# Two Gamma Dist
data2 = np.concatenate([stats.gamma(5).rvs(50) - 1,
                        -1*stats.gamma(5).rvs(50)])
```

```python
# 只看boxplot這兩個dataset很相似
sns.boxplot(data=[data1, data2], whis=np.inf)
```

```python
# 看violin plot可以看出這兩個dataset的密度分佈差很多
sns.violinplot(data=[data1, data2])
```

```python
# 試著調整一些attributes
sns.violinplot(data2, bw=0.01, orient='v')
```

```python
# 類似rug plot，用線條展現data的密度
sns.violinplot(data1, inner='stick', orient='v')
```
