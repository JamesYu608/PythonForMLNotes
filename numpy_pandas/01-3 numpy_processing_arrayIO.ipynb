# Array Processing

```python
import numpy as np
# 將資料視覺化，之後會再看更詳細
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
points = np.arange(-5, 5, 0.001)
points
```

這邊公式內容不重要，只是為了產生資料來示範資料視覺化

```python
# 產生網格矩陣
dx, dy = np.meshgrid(points, points)
```

```python
z = (np.sin(dx) + np.sin(dy))
z
```

```python
# 顯示視覺化資料
plt.imshow(z)
plt.colorbar()
plt.title('Plot for sin(x) + sin(y)')
```

## numpy where

```python
A = np.array([1, 2, 3, 4])
B = np.array([100, 200, 300 ,400])
```

```python
condition = np.array([True, True, False, False])
```

```python
[(A_val if cond else B_val) for A_val, B_val, cond in zip(A, B, condition)]
```

```python
# 使用numpy的where來做到和上面一樣的事情
np.where(condition, A, B)
```

Example:

```python
from numpy.random import randn
```

```python
arr = randn(5, 5)
arr
```

```python
np.where(arr < 0, 0, arr)
```

## Other Statistical Processing

```python
arr = np.arange(1, 10).reshape([3, 3])
arr
```

```python
# Sum
arr.sum()
```

```python
# Sum with axis
arr.sum(0) # across - 0: rows, 1: columns
```

```python
# Mean
arr.mean()
```

```python
# Variance
arr.var()
```

```python
# Standard Deviation
arr.std()
```

```python
arr = np.array([3, 2, 1])
# Sort
arr.sort()
arr
```

```python
bool_arr = np.array([True, False, False])
```

```python
# 只要有一個True就為True
bool_arr.any()
```

```python
# 全部都為True才為True
bool_arr.all()
```

```python
countries = np.array(['France', 'Germany', 'USA', 'Russia', 'USA', 'Mexico', 'Germany'])
# Unique
np.unique(countries)
```

```python
# 是否有在1-D array裡面
np.in1d(['France','USA','Sweden'], countries)
```

# Array Input and Output

```python
import numpy as np
```

```python
arr = np.arange(5)
arr
```

```python
# 存成myarray.npy
np.save('myarray', arr)
```

```python
arr = np.arange(10)
arr
```

```python
# 讀回來
arr = np.load('myarray.npy')
arr
```

```python
arr2 = np.arange(10)
arr2
```

```python
# 打包多個array，存成.npz檔案
np.savez('ziparray.npz', x=arr, y=arr2)
```

```python
# 讀回來
archive_array = np.load('ziparray.npz')
```

```python
archive_array['x']
```

```python
archive_array['y']
```

```python
arr = np.arange(10).reshape([2, 5])
arr
```

```python
# 存成.txt
np.savetxt('mytextarray.txt', arr, delimiter=',')
```

```python
# 讀回來 (delimiter要一樣)
arr = np.loadtxt('mytextarray.txt', delimiter=',')
arr
```
