# Histograms
## `matplotlib.pyplot.hist`

```python
dataset1 = randn(1000)
plt.hist(dataset1)
```

```python
# Set colors (可以查看mpl.colors.cnames)
dataset2 = randn(1000)
plt.hist(dataset2, color='indianred')
```

```python
# normed - 以機率分佈顯示
# alpha - 透明度(0~1)
# bins - 分成多少部分
plt.hist(dataset1, normed=True, color='indianred', alpha=0.4, bins=20)
plt.hist(dataset2, normed=True, color='blue', alpha=0.4, bins=20)
```

## `seaborn.joinplot`

```python
data1 = randn(1000)
data2 = randn(1000)
# 上方和右方是單變量的直方圖
# 中間是雙變量的散布圖
# 兩個dataset的elements數量要一樣
sns.jointplot(data1, data2)
```

```python
# Set kind (scatter | reg | resid | kde | hex)
# hex是hexagon的縮寫
sns.jointplot(data1, data2, kind='hex')
```

```python
sns.jointplot(np.array([2, 1, 3]), np.array([1, 4, 9]))
```
